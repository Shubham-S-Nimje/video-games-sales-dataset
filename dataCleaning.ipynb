```python
import numpy as np 
import pandas as pd

data = pd.read_csv(r'D:\Learning\numpy\Video Games\VideoGamesSales.csv')
data.shape
```

```python
data.head(2)
```

```python
data[data.duplicated()].shape # Check for duplicates
```

```python
data.drop_duplicates(inplace=True) # Remove duplicates
```

```python
data[data.duplicated()].shape # Check for duplicates
```

```python
data.isnull().sum() # Check for null values
```

```python
# data[data['Publisher'].isnull()]
data['Publisher'].isnull().sum() # Check for null values in 'Publisher' column
```

```python
data['Publisher'].fillna('Unknown', inplace=True) # Fill null values in 'Publisher' with 'Unknown'
```

```python
data['Publisher'].isnull().sum()
```

```python
data.isnull().sum()
```

```python
data['Region'].isnull().sum()
```

```python
data['Region'].fillna('North', inplace=True) # Fill null values in 'Region' with 0
```

```python
data.isnull().sum()
```

```python
data['NA_Sales'].replace('[$]', '', regex=True, inplace=True) # Remove $ sign from 'NA_Sales'
data 
```

```python
pd.to_numeric(data['NA_Sales'], errors='coerce') # Convert 'Global_Sales' to numeric, coercing errors
```

```python
# Ensure NA_Sales is numeric before calculating mean
data['NA_Sales'] = pd.to_numeric(data['NA_Sales'], errors='coerce')
avaraege_sales = data['NA_Sales'].mean() # Calculate average sales
int(avaraege_sales)
```

```python
data['NA_Sales'] = data['NA_Sales'].fillna(avaraege_sales) # Fill null values in 'NA_Sales' with average sales
data
```

```python
data['Country'] = data['Country'].str.replace('United States', 'USA') # Replace 'United States' with 'USA'
data
```

```python
data[data['Country'] == 'USA'].shape # Check how many rows have 'Country' as 'USA'
```

```python
data.rename(columns={'NA_Sales': 'National Sales', 'Global_Sales': 'Global Sales'}, inplace=True) # Rename columns
data
```

```python
data.rename(columns={'NA_Profit': 'National Profit', 'Global_Profit': 'Global Profit'}, inplace=True) # Rename columns
data
```

```python
sales_cap = data['National Sales'].quantile(0.95) # Calculate 95th percentile for 'National Sales'
int(sales_cap)
```

```python
data['National Sales'] = np.where(data['National Sales'] > sales_cap, sales_cap, data['National Sales']) # Cap 'National Sales' at 95th percentile
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.head(2)
```

```python
pd.to_numeric(data['National Sales'], errors='coerce')
```

### Create bar chart

```python
National_Sales = data.groupby(['Region','Country'])['National Sales'].sum().reset_index().sort_values(by='National Sales', ascending=True)
National_Sales
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='Region', y='National Sales', data=National_Sales, hue='Country')

plt.title('National Sales by Region and Country', fontweight='bold', fontsize=16)
plt.xlabel('Region', fontweight='bold', fontsize=16)
plt.ylabel('National Sales', fontweight='bold', fontsize=16)    
plt.show()
```

### Create bax chart

```python
data.head(2)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='National Sales', data=data, showmeans=True, hue='Genre', meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"green", "markersize":"10"}) 
plt.title('National Sales Distribution by Region and Country', fontweight='bold', fontsize=16)
plt.xlabel('Region', fontweight='bold', fontsize=16)    
plt.ylabel('National Sales', fontweight='bold', fontsize=16)
plt.legend(title='Country', loc='upper right')
plt.show()
```

```python
data.head(2)
```

### Create pie chart

```python
Sales = data.groupby(['Country'])[['National Sales', 'Global Sales']].sum().reset_index()
Country = Sales['Country']
National_Sales = Sales['National Sales']
Global_Sales = Sales['Global Sales']

fig, axs = plt.subplots(1, 2, figsize=(10, 7))

axs[0].pie(National_Sales, labels=Country, autopct='%1.1f%%', startangle=90)
axs[0].set_title('National Sales by Country', fontweight='bold', fontsize=16)

axs[1].pie(Global_Sales, labels=Country, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Global Sales by Country', fontweight='bold', fontsize=16)

plt.show()
```

### Create line chart

```python
#create a line chart 
plt.figure(figsize=(10, 6))

#Line chart for National Sales
plt.plot(data['Year'], data['National Sales'], marker='o', linestyle='-', color='b', label='National Sales')

#Line chart for Global Sales
plt.plot(data['Year'], data['Global Sales'], marker='s', linestyle='--', color='g', label='Global Sales')

#adding title and labels
plt.title('National Sales and Global Sales Over Years', fontweight='bold', fontsize=16)
plt.xlabel('Year', fontweight='bold', fontsize=16)
plt.ylabel('Sales', fontweight='bold', fontsize=16)
plt.legend()

# Display the line chart
plt.grid(True)
plt.show()
```
